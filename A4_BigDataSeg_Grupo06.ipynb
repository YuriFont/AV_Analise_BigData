## Criando link com o dataset e mostrando em formato de Tabela

```python
import pandas as pd
tabela = pd.read_excel('Stores.xlsx')
display(tabela)
```

## Inserindo Coluna
<p>
    Criamos um função "imposto" que irá diminuir em 10% o valor que foi passao como parâmetro, para simular caso quiséssemos aplicar a taxação de um imposto em um relatório.
    Depois pegamos a mesma tabela que foi criada a cima e adcionamos a ela uma nova coluna chamada "Tax" para assim simular quanto pagamos de imposto.
    Logo abaixo criamos outra função que irá nos retornar o valor do 'Store_Sales' menos quanto iremos pagar de imposto. E assim criando outra coluna chamada "Post_Tax_Sales".
</p>

```python
def imposto(valor):
    return round(valor * 0.10, 2)
def cal_imposto(valor):
    return round(valor - imposto(valor))
tabela['Tax'] = tabela['Store_Sales'].map(imposto)
tabela['Post_Tax_Sales'] = tabela['Store_Sales'].map(cal_imposto)
display(tabela)
```

## Ordenando Coluna

<p>
   Aqui nós chamamos mais uma vez o dataset para ser manipulado, criando uma nova variável que irá receber esse datatset porém com apenas as colunas "ID" e "Store_Sales". Após isso nós utilizamos da método 'sort_values' para ordenar em ordem decrescente apenas a coluna 'Store_Sales'.  
</p>

```python
tabela_total = pd.read_excel('Stores.xlsx')
tabela_ordenada = tabela_total[['ID', 'Store_Sales']]
tabela_ordenada = tabela_ordenada.sort_values(by='Store_Sales', ascending=False)
display(tabela_ordenada)
```

## Trabalhando com calculos estatíticos

<p>
     Criamos três diferentes variáveis para calcular a Média, Mediana e Moda. Depois criamos o DataFrame de cada variável para assim podermos transformá-las em uma tabela, onde elas seriam as colunas e as linhas seriam os informativos da tabela.
    
    OBS: Tivemos que utilizar o método '.iloc()' na variável 'modas' pois o método '.mode()' cria um array de modas, já que dentro de um datatset não existem apenas um tipo de valor que se repete.
</p>

```python
colunas_selecionadas = ['Store_Area', 'Items_Available', 'Daily_Customer_Count', 'Store_Sales']

medias = tabela_total[colunas_selecionadas].mean()
medianas = tabela_total[colunas_selecionadas].median()
modas = tabela_total[colunas_selecionadas].mode().iloc[0]

df_medias = pd.DataFrame({'Médias': medias})
df_medianas = pd.DataFrame({'Medianas': medianas})
df_modas = pd.DataFrame({'Modas': modas})

tabela_medias_medianas_modas = pd.concat([df_medias, df_medianas, df_modas], axis=1)

display(tabela_medias_medianas_modas)
```

## Problema identificado
<br>
<p>Ao analisar o nosso dataset identificamos que existem algumas lojas com uma discrepância de vendas, como exemplo as lojas de ID 868 e 32 respectivamente. Nelas nos podemos verificar que mesmo ambas tendo uma área e quantidade de produtos bem parecidas, ambas tiveram a mesma quantidade clientes no mês porém com valores de vendas bem distintos.</p><br><p> Podemos assim considerar a média de vendas da loja ID 32 igual a 15,07 enquanto a loja ID 869 teve 107,29. Um dos problemas que podemos observar é a quantidade de itens que a loja 868 tem a mais que a outra, isso pode acarretar em uma diversidade de itens mais abrangente trazendo assim mais clientes. Isso pode ser ocasionado por conta do espaço que a loja 32 tem (1250), o que leva a mesma a ter um estoque reduzido em relação a outra.</p>
<br>

```python
tabela_total = pd.read_excel('Stores.xlsx')

colunas_selecionadas = ['ID', 'Store_Area', 'Items_Available', 'Daily_Customer_Count', 'Store_Sales']

primeira_linha = tabela_total[colunas_selecionadas].iloc[868]
ultima_linha = tabela_total[colunas_selecionadas].iloc[31]

tabela_resultado = pd.DataFrame([primeira_linha, ultima_linha])

display(tabela_resultado)
```
